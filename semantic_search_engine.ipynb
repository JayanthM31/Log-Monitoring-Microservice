```python
# Install Flask
!pip install flask

from flask import Flask, jsonify

# Step 1: Sample logs
logs = [
    {"level": "INFO", "message": "System started successfully"},
    {"level": "ERROR", "message": "Database connection failed"},
    {"level": "WARNING", "message": "High memory usage detected"},
    {"level": "INFO", "message": "User logged in"},
    {"level": "ERROR", "message": "API timeout error"},
    {"level": "INFO", "message": "Service running normally"}
]

# Step 2: Create Flask app
app = Flask(__name__)

@app.route("/")
def home():
    return "Log Monitoring Microservice Running!"

@app.route("/logs")
def get_logs():
    return jsonify(logs)

@app.route("/errors")
def get_errors():
    return jsonify([log for log in logs if log["level"] == "ERROR"])

@app.route("/warnings")
def get_warnings():
    return jsonify([log for log in logs if log["level"] == "WARNING"])

# Step 3: Run on different port (fix error)
app.run(port=5001)
```
